```python
#PRIMERA VERSION DEL SISTEMA EXPERTO

from experta import *

class Paciente(Fact):
  """Information about the paciente"""
  pass

class MentalHealthExpert(KnowledgeEngine):

  @DefFacts()
  def _initial_action ( self ) :
    yield Fact(nombre="Luis")
    yield Fact(edad="22")
    yield Fact(genero="hombre")
    yield Fact(id="123456")
    yield Fact(sangre="o-")


  #@Rule(AS.myfact << Fact(W()))
  @Rule(OR(Paciente(s2='trouble.concentrating')))
  def stressed(self):
      evidence = {'trouble.concentrating': 'yes'}
      print(Paciente['s1'])
      print('PROBABILITIY GIVEN SIMPTOMS')
      prob = infer.query(variables=['Disorder'], evidence=evidence)
      print(prob)
      print("Practice stress-reduction techniques such as deep breathing, yoga, or tai chi.")
      print("Take regular breaks throughout the day to relax and recharge.")
      print("Prioritize tasks and delegate when possible to reduce feelings of overwhelm.")

  @Rule(OR(Paciente(s1='panic')))
  def Anxiety(self):
    evidence = {'panic': 'yes'}
    print('PROBABILITIY GIVEN SIMPTOMS')
    prob = infer.query(variables=['Disorder'], evidence=evidence)
    print(prob)
    print("Practice deep breathing exercises and mindfulness meditation.")
    print("Engage in regular physical activity and maintain a healthy diet.")
    print("Limit caffeine and alcohol intake, as they can exacerbate anxiety symptoms.")

  @Rule(Paciente(s1='feeling.tired'))
  def lonely(self):
    evidence = {'feeling.tired': 'yes'}
    print('PROBABILITIY GIVEN SIMPTOMS')
    prob = infer.query(variables=['Disorder'], evidence=evidence)
    print(prob)
    print("Seek support from loved ones or a therapist.")
    print("Join social activities to meet new people.")
    print("Practice self-compassion and engage in hobbies.")

  @Rule(Paciente(s2='hopelessness'), salience=1)
  def depressed(self):
    evidence = {'hopelessness': 'yes'}
    print('PROBABILITIY GIVEN SIMPTOMS')
    prob = infer.query(variables=['Disorder'], evidence=evidence)
    print(prob)
    print("Seek support from friends, family, or a therapist to talk about your feelings.")
    print("Engage in activities you enjoy, even if you don't feel like it at first.")
    print("Establish a daily routine and set achievable goals to regain a sense of control.")

  @Rule(AND(Paciente(s1="panic"), Paciente(s2="hopelessness")), salience=1)
  def anxiety_and_stress(self):
      print("Combine stress-reduction techniques with anxiety management strategies.")
      print("Practice self-compassion and accept that it's okay to ask for help when needed.")

  @Rule(AND(Paciente(s1="feeling.tired"), Paciente(s2="hopelessness")))
  def feeling_tired_and_hopelessness(self):
    print("Prioritize self-care and seek professional support.")
    print("Exercise and healthy habits can improve energy levels.")
    print("Therapy can help address feelings of hopelessness and negative thoughts.")

  @Rule(Paciente(s2="suicidal.thought"))
  def suicidal_thought(self):
    print("Seek immediate help from a mental health professional or a trusted person.")
    print("Remove access to any means of self-harm, such as firearms or medications.")
    print("Create a safety plan outlining coping strategies and emergency contacts.")

  @Rule(Paciente(s2="avoids.people.or.activities"))
  def avoids_people_or_activities(self):
    print("Gradually expose yourself to social situations or activities you've been avoiding.")
    print("Start with small steps and reward yourself for each accomplishment.")
    print("Consider seeking support from a therapist to address underlying causes of avoidance.")

  @Rule(AND(Paciente(s2="trouble.concentrating"), Paciente(s1="feeling.tired")))
  def trouble_concentrating_and_feeling_tired(self):
    print("Prioritize rest and relaxation to improve concentration.")
    print("Break tasks into smaller, manageable steps to avoid feeling overwhelmed.")
    print("Engage in activities that boost energy levels and mental clarity.")


  @Rule(OR(Paciente(s1=W()),Paciente(s2=W())), salience=0)
  def default(delf):
    print()
    print('GENERAL STATISTICS')
    prob = infer.query(variables=['Disorder'])
    print(prob)
    pass


#LAS ENTRADAS SON DE 1 A 2 SINTOMAS, s1 y s2 respectivamente

engine = MentalHealthExpert()
engine.reset()


#ESTA ES LA ENTRADA DEL SISTEMA
#los posibles valores para s1 son: 'panic', 'anger', 'none', 'feeling.tired'
#los posibles valores para s2 son: 'suicidal.thought', 'hopelessness', 'avoids.people.or.activities', 'trouble.concentrating'
engine.declare(Paciente(s1='panic', s2='hopelessness'))

engine.run()  # Run the expert system
```
